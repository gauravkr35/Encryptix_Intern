```python
import numpy as np
```

```python
Q1=np.percentile(dataset["sepal_width"],25)
Q3=np.percentile(dataset["sepal_width"],75)
```

```python
Q1,Q3
```

```python
IQR=Q3-Q1
min_outliers=Q1-1.5*IQR
max_outliers=Q3+1.5*IQR
IQR,min_outliers,max_outliers
```

```python
(dataset['sepal_width']<min_outliers).values.sum()
```

```python
(dataset['sepal_width']>max_outliers).values.sum()
```

```python
dataset=dataset[(dataset['sepal_width']>min_outliers) & (dataset['sepal_width']<max_outliers)]
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(3,2))
sns.boxplot(x="sepal_width",data=dataset)
plt.show()
```

```python
dataset['species'].value_counts()
```

```python
dataset.replace('Iris-virginica',0,inplace=True)
dataset.replace('Iris-versicolor',1,inplace=True)
dataset.replace('Iris-setosa',2,inplace=True)
```

```python
dataset=dataset.drop_duplicates() 
```

```python
dataset.info()
```

```python
sns.pairplot(data=dataset , hue='species')
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

```python
tree=DecisionTreeClassifier(max_depth=4)
```

```python
tree.fit(x_train,y_train)
```

```python
tree.score(x_test,y_test)*100
```

```python
tree.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(10,10))
plot_tree(tree)
plt.savefig("graph.jpg")
plt.show()
```

```python
for i in range(1,20):
    t2=DecisionTreeClassifier(max_depth=i)
    t2.fit(x_train,y_train)
    print(t2.score(x_test,y_test)*100,t2.score(x_train,y_train)*100,  i)
```

```python
### max_depth = 4
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf2=confusion_matrix(y_test,tree.predict(x_test))
cf2
```

```python
sns.heatmap(cf2,annot=True)
plt.show()
```
